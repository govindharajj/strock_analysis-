```python
import os
import yaml
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Define the folder containing YAML files
base_folder = r"D:\stock analysis\data"

# List to store the loaded data
all_yaml_data = []

# Walk through all directories and files in the base folder
for root, dirs, files in os.walk(base_folder):
    
    for file in files:
        if file.endswith(".yaml"):  # Check for YAML files
            yaml_file_path = os.path.join(root, file)
           
            # Read and load the YAML file
            try:
                with open(yaml_file_path, "r") as f:
                    yaml_data = yaml.safe_load(f)  # Load YAML file as a Python dictionary
                    all_yaml_data.append(yaml_data)  # Add the data to the list

            except Exception as e:
                pass
```

```python
final = []

for i in all_yaml_data:
    for j in i:
        final.append(j)
df1 = pd.DataFrame(final)

df1.head()
```

```python


# ✅ Step 1: All Nifty 50 Tickers
tickers = [
    'SBIN', 'BAJFINANCE', 'TITAN', 'ITC', 'TCS', 'LT', 'TATACONSUM', 'RELIANCE',
    'HCLTECH', 'JSWSTEEL', 'ULTRACEMCO', 'POWERGRID', 'INFY', 'TRENT', 'BHARTIARTL',
    'TATAMOTORS', 'WIPRO', 'TECHM', 'NTPC', 'HINDUNILVR', 'APOLLOHOSP', 'M&M',
    'GRASIM', 'ICICIBANK', 'ADANIENT', 'ADANIPORTS', 'BEL', 'BAJAJFINSV',
    'EICHERMOT', 'COALINDIA', 'MARUTI', 'INDUSINDBK', 'ASIANPAINT', 'TATASTEEL',
    'HDFCLIFE', 'DRREDDY', 'SUNPHARMA', 'KOTAKBANK', 'SHRIRAMFIN', 'NESTLEIND',
    'ONGC', 'CIPLA', 'BPCL', 'BRITANNIA', 'SBILIFE', 'HINDALCO', 'HEROMOTOCO',
    'AXISBANK', 'HDFCBANK', 'BAJAJ-AUTO'
]

# ✅ Step 2: Random Open & Close Prices
np.random.seed(42)
open_prices = np.random.randint(500, 4000, size=len(tickers))
close_prices = open_prices + np.random.randint(-100, 100, size=len(tickers))

# ✅ Step 3: Calculate Daily Return (%)
daily_return = np.round(((close_prices - open_prices) / open_prices) * 100, 2)

# ✅ Step 4: Create DataFrame
df4 = pd.DataFrame({
    "Ticker": tickers,
    "open": open_prices,
    "close": close_prices,
    "daily_return": daily_return
})

df4["daily_return"] = df4["close"] - df4["open"]
df4["daily_return_pct"] = np.round(((df4["close"] - df4["open"]) / df4["open"]) * 100, 2)
# ✅ Step 5: Print full DataFrame
print(" Nifty 50 Daily Stock Data:\n")
print(df4.to_string(index=False))
#filtered_df = df4[df4["Ticker"] == "SBIN"][["Ticker", "open", "close", "daily_return","daily_return_pct"]]
#print(filtered_df)

# ✅ Step 6 (Optional): Loop through each Ticker and show individual data
print("\n Individual Ticker Breakdown:")
for ticker in df4["Ticker"].unique():
    print("\n" + "="*40)
    print(f"🔸 Ticker: {ticker}")
    print(df4[df4["Ticker"] == ticker][["Ticker", "open", "close", "daily_return"]].to_string(index=False))
```

```python
# Establish the connection
connection = mysql.connector.connect(
        host="127.0.0.1",
        port=3306,
        user='root',
        password="root",
       database="stock_analysis" )
cursor = connection.cursor()
```

```python
cursor.execute("""
    CREATE TABLE IF NOT EXISTS stock_analysis (
        Ticker   VARCHAR(50),
        close    VARCHAR(50),
        date     DATE,
        high     VARCHAR(50),
        low      VARCHAR(50),
        month      VARCHAR(50),  
        open     VARCHAR(50),
        volume    VARCHAR(50),
        sector   VARCHAR(100),
        symbol   VARCHAR(50),      
        daily_return VARCHAR(50),
        Daily_return_pct VARCHAR(50)
    ); 
""")
```

```python


insert_query = """
INSERT INTO stock_analysis(Ticker,Close, date, high, low, month, open, volume, sector, symbol, daily_return, Daily_return_pct)  
VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
"""

for index, row in df3.iterrows():
    values = (
        row['ticker'],
        row[' close '],
        row['date'],
        row['High'],
        row['low'],
        row['Month'],
        row['open'],
        row['Volume'],
        row['sector'],
        row['Symbol'],  
        row['daily_return'],
        row['Daily_return_pct']
    )
    
    cursor.execute(insert_query, values)
    connection.commit()
    try:
        cursor.execute(insert_query, values)
    except Exception as e:
        print(f"❌ Row {index} error: {e}\nValues: {values}")
```
